```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Product": [
        "Laptop",
        "Phone",
        "Tablet",
        "Monitor",
        "Keyboard"
    ],
    "Sales": [
        1200,
        900,
        500,
        300,
        150
    ],
    "Region": [
        "North",
        "South",
        "North",
        "West",
        "East"
    ]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Product'],
       df['Sales']
       )
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title("Sales by product")
plt.xticks(rotation=0)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(
    df["Sales"],
    bins=30,
    kde=True
)

plt.title("Sales Distribution")

plt.show()
```

```python
category_sales = (
    df.groupby("Product")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

category_sales
```

```python
plt.figure(figsize=(8,5))

category_sales.plot(
    kind="bar"
)

plt.title("Sales by Category")

plt.ylabel("Total Sales")

plt.show()
```
