```python
import pandas as pd

data = {
    "Product": [
        "Laptop",
        "Phone",
        "Tablet",
        "Monitor",
        "Keyboard"
    ],
    "Sales": [
        1200,
        900,
        None,
        300,
        150
    ],
    "Region": [
        "North",
        "South",
        "North",
        "West",
        "East"
    ]
}

df = pd.DataFrame(data)

df
```

```python
df.info()
df.describe()

df.isnull().sum()
```

```python
df['Sales'] = df["Sales"].fillna(df['Sales'].median())
```
